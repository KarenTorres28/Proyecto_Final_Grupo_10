```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
base = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
def suma(x):
    s = 0
    x_str = str(x)
    x_limpia = x_str.replace('.', '')
    for i in x_limpia:
        if i.isdigit():
            s += int(i)
    return s
base.apply(lambda x: suma(x.iat[0]), axis = 1)
```

```python
base["n"] =base.apply(lambda x: suma(x.iat[0]), axis = 1)
base.head()
```

```python
def comp(x,y):
    if x == y:
        return 5
    else:
        return 1
```

```python
def costo(n,x):
    resto = 2*x + 23
    valorc = n*x**2
    resul = resto + valorc
    return resul
base.iloc[:, 1:] = base.iloc[:, 1:].astype(str).replace(',', '.', regex=True).apply(pd.to_numeric, errors='coerce').round(2)
base["Punto 2"] = base.apply(lambda x: (costo(x.iat[4],101)-costo(x.iat[4],100))/1 , axis = 1 )
base["CP2"] = base.apply(lambda x: comp(x.iat[2],x.iat[7]), axis = 1)
base.head()
```

```python
base.iloc[:, 1:] = base.iloc[:, 1:].astype(str).replace(',', '.', regex=True).apply(pd.to_numeric, errors='coerce').round(2)
base["Punto 3"] = base.apply(lambda x: (2 * x.iat[4] * 100)+2,  axis=1).round(2)
base["CP3"]  = base.apply(lambda x: comp(x.iat[3],x.iat[9]), axis = 1) 
base.head()
```

```python
def calificacion_final(c1,c2,c3):
    s = c1+c2+c3
    res = s/3
    return res
```

```python
base["NF "] = base.apply(lambda x :calificacion_final(x.iat[6],x.iat[8],x.iat[10]), axis = 1)
base
```
