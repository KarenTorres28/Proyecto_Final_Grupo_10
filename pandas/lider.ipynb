```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
base = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
def suma(x):
    s = 0
    x_str = str(x)
    x_limpia = x_str.replace('.', '')
    for i in x_limpia:
        if i.isdigit():
            s += int(i)
    return s
base.apply(lambda x: suma(x.iat[0]), axis = 1)
```

```python
base["n"] =base.apply(lambda x: suma(x.iat[0]), axis = 1)
base.head()
```

```python
def comp(x,y):
    if x == y:
        return 5
    else:
        return 1
```

```python
def costo(n,x):
    resto = 2*x + 23
    valorc = n*x**2
    resul = resto + valorc
    return resul
base.iloc[:, 1:] = base.iloc[:, 1:].astype(str).replace(',', '.', regex=True).apply(pd.to_numeric, errors='coerce').round(2)
base["Punto 1"] = base.apply(lambda x: (costo(x.iat[4],105)-costo(x.iat[4],100))/5 , axis = 1 ) #Convierte toso los datos a numeros, primero convierte todo a str temporalmente, despues reemplaza comas por puntos y convierte lo corregido a numerico
base["CP1"] = base.apply(lambda x: comp(x.iat[1],x.iat[5]), axis = 1)
base.head()
```
